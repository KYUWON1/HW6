```python
import csv 
import matplotlib.pyplot as plt

def main():

    print("Hello, World!")

if __name__ == '__main__':

 

    main()

f = open('subway.csv')
data = csv.reader(f)
next(data)
next(data)

n = 30
station_data = []

for row in data:
    station = row[1] + " " + row[3]  #    
    on = int(row[4].replace(',', ''))+int(row[6].replace(',', ''))
    off = int(row[5].replace(',', ''))+int(row[7].replace(',', ''))
    onoff = int(row[4].replace(',', '')) + int(row[5].replace(',', ''))+int(row[6].replace(',', ''))+int(row[7].replace(',', ''))
    station_data.append((station, on, off, onoff))

#print(station_data)
station_data.sort(key=lambda x: x[1], reverse=True)
on_top_n = [x[1] for x in station_data[:n]]
on_top_n_stations = [x[0] for x in station_data[:n]]

station_data.sort(key=lambda x: x[2], reverse=True)
off_top_n = [x[2] for x in station_data[:n]]
off_top_n_stations = [x[0] for x in station_data[:n]]

station_data.sort(key=lambda x: x[3], reverse=True)
onoff_top_n = [x[3] for x in station_data[:n]]
onoff_top_n_stations = [x[0] for x in station_data[:n]]

plt.figure(figsize=(12, 8))

plt.rc('font',family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False


plt.subplot(1, 3, 1)
plt.barh(on_top_n_stations, on_top_n)
plt.xlabel(' ')
plt.title('   {} '.format(n))
plt.xticks(range(0, 800000, 200000))

plt.subplot(1, 3, 2)
plt.barh(off_top_n_stations, off_top_n)
plt.xlabel(' ')
plt.title('   {} '.format(n))
plt.xticks(range(0, 800000, 200000))

plt.subplot(1, 3, 3)
plt.barh(onoff_top_n_stations, onoff_top_n)
plt.xlabel(' ')
plt.title('   {} '.format(n))
plt.xticks(range(0, 800000, 200000))

plt.tight_layout()
plt.show()
```
