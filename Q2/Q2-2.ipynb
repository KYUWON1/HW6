```python
import csv
import matplotlib.pyplot as plt

def main():

    print("Hello, World!")


if __name__ == '__main__':

    main()
    
f = open('result.csv','r',encoding='ANSI')
data = csv.reader(f)
next(data)


data1 = []
data2 = []
data3 = []
data4 = []


for row in data:
    if row[-1] != '':
        for i in range(2,8):
            if row[0] == '100':
                data1.append(int(row[i]))
            if row[0] == '1000':
                data2.append(int(row[i]))
            if row[0] == '10000':
                data3.append(int(row[i]))
            if row[0] == '100000':
                data4.append(int(row[i]))
                
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
axs[0, 0].bar([1, 2, 3, 4, 5, 6], data1, color='b')
axs[0, 0].set_title('100 result')
axs[0, 1].bar([1, 2, 3, 4, 5, 6], data2, color='r')
axs[0, 1].set_title('1000 result')
axs[1, 0].bar([1, 2, 3, 4, 5, 6], data3, color='g')
axs[1, 0].set_title('10000 result')
axs[1, 1].bar([1, 2, 3, 4, 5, 6], data4, color='purple')
axs[1, 1].set_title('100000 result')

for ax in axs.flat:
    ax.set(xlabel='dice number', ylabel='count')

plt.show()
```

해당 결과를 보았을때, 주사위를 100번, 1000번 던졌을땐, 그래프간의 간격의 차이가 많이 나지만, 실행 횟수가 늘어남에 따라 그래프가 평평해지는 것을 볼 수있습니다. 따라서, 주사위를 많이 던질수록, 각각 숫자가 등장할 확률이 균등해지고, 평균값은 평균 값인 3.5에 가까워 집니다.
